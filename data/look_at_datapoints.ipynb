```python
import json
from collections import defaultdict, Counter
with open("dataset_en_split_dev.json", "r", encoding="utf-8") as injson:
    data = json.load(injson)

text_len = defaultdict()
for i, datapoint in enumerate(data):
    text_len[i] = len(datapoint['text'].split())

sorted(text_len.items(), key=lambda x: x[1], reverse=True)
```

```python
import random
def sample_data(data:dict, seed=42, k=10):
    ratio = 1/3
    conspiracy = [datapoint for datapoint in data if datapoint["category"] == "CONSPIRACY"]
    critical = [datapoint for datapoint in data if datapoint["category"] == "CRITICAL" ]

    random.seed(seed)
    conspiracy = random.sample(conspiracy, round(k*ratio))
    critical = random.sample(critical, round((1-ratio)*k))
    return conspiracy + critical

def sample_toy_data(lang):
    filename = f"dataset_{lang}_train.json"
    with open(filename, "r", encoding="utf-8") as injson:
        data = json.load(injson)
    toy_sample = sample_data(data)
    new_filename = f"dataset_{lang}_split_toy.json"
    with open(new_filename, "w", encoding="utf-8") as outjson:
        json.dump(toy_sample, outjson)
```

```python
sample_toy_data("en")
sample_toy_data("es")
```
